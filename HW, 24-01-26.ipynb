#Difference between is and ==
"is" used only to compare between two variables
"==" used to compare between variables and any data types

```python
4 is 6
```

```python
"a" is "b"
```

```python
a=4
b=5
a is b
```

```python
4==7
```

```python
a==b
```

```python
"a"=="A"
```

```python
#explore is, and, or, not, in, +=,-=
print(12 is 12) #--- True
print(4>3 and 4>5) #---False
print(4>3 or 4>5) #---True
a="hello"
b="helloworld"
print(a in b) #---True
x=8
x+=4 # x=x+4 === x=8+4 == 12
print(x)
y=2 
y-=6 #y=y-6 === y=2-6 == -4
print(y)
```
