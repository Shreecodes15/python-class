1. Print numbers from 1 to 10 except 3 and 8
2. Find the first number divisible by 7 between 1 and 100
4. Count vowels in a string
5. Print all numbers from 1 to 30 except multiples of 5
6. Print numbers 1–50 but skip multiples of 3, skip numbers ending with 7, stop completely if number becomes 40

```python
#Print numbers from 1 to 10 except 3 and 8
i=1
while i<=10:
    if i==3 or i ==8:
        i=i+1
        continue
    print(i)
    i=i+1
```

```python
#Find the first number divisible by 7 between 1 and 100
i=1
while i<=100:
    if i%7==0:
        print("first Number divisible by 7:",i)
        i=i+1
        break
    print(i)
    i=i+1
```

```python
#Count vowels in a string
s=input("enter your string: ")
l=s.lower()
for i in l:
    if i in "aeiou":
        print("vowel--",i)
```

```python
#Print all numbers from 1 to 30 except multiples of 5
i=1
while i<=30:
    if i%5==0:
        i=i+1
        print("_",end=" ")
        continue
    print(i,end=" ")
    i=i+1
```

```python
#Print numbers 1–50 but skip multiples of 3, skip numbers ending with 7, stop completely if number becomes 40
i=1
while i<=50:
    if i%3==0 or i%10==7:
        i=i+1
        print("*", end=" ")
        continue
    elif i==40:
        i=i+1
        print("^^^end of program^^^")
        break
    else:
        print(i,end=" ")
    i=i+1
```
