```python
#print square root of the number
import math
n=float(input("Enter your number "))
if n<0:
    print("Invalid number")
else:
    s=math.sqrt(n)
    print("square root of {} is {}".format(n,s))
```

```python
#Nested if and if-elif-else : Largest of Three Numbers
n1=int(input("enter number one"))
n2=int(input("enter number two"))
n3=int(input("enter number three"))
if n1>n2 and n1>n3:
        print("Largest number among three is ",n1)
elif n2>n3 and n2>n1:
        print("Largest number among three is ",n2)
elif n3>n1 and n3>n2:
        print("Largest number among three is ",n3)
```

```python
#Write a program to print numbers from 1 to 10 using for loop.
for i in range(1,11):
    print(i)
```

```python
#Write a program to find the sum of numbers from 1 to 10 using a loop.
s=0
for i in range(1,11):
    s=s+i
print('addition=',s)
```

```python
 #Write a program that prints the numbers from 10 to 1 using a for loop.
for i in range(10,0,-1):
    print(i)
```

```python
#Write a program that print the square of each number from 1 to 10.
for i in range(1,11):
    print("square of {} is {}".format(i,i*i))
```

```python
#Write a program to display the multiplication table of a number entered by the user.
#create table by user input number
n=int(input("enter number to print its table "))
for i in range(1,11):
    print("{}X{}={}".format(n,i,n*i))
```
